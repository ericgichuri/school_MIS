```python
im=Image.open('icons/ESSLOGO.png').resize((169,174))
bgsize=((im.size[0]*3)+10,(im.size[1]*3)+10)
mask=Image.new('L',bgsize,0)
#mask=Image.new('RGBA',bgsize,(0,0,0))
draw=ImageDraw.Draw(mask)
draw.ellipse((0,0)+bgsize,fill=255)
mask=mask.resize(im.size)
im.putalpha(mask)
im
output=ImageOps.fit(im,mask.size,centering=(0.5,0.5))
output.putalpha(mask)
output.save('output.png')
template.paste(im,(0,0))
template
```
