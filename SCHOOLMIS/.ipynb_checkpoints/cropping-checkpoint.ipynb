```python
from PIL import Image,ImageTk,ImageOps,ImageDraw,ImageFilter

im=Image.open('students/student1.jpg')
bgsize=(im.size[0]*3,im.size[1]*3)
mask=Image.new('L',bgsize,0)
draw=ImageDraw.Draw(mask)
draw.ellipse((0,0)+bgsize,fill=255)
mask=mask.filter(ImageFilter.GaussianBlur)
mask=mask.resize(im.size)
im.putalpha(mask)
im
```
